```python
import math

#  1
def calculate_w(x: float, y: float):
    try:
        w = (x + y) / (y + 1) - (x * y - 12) / (34 + x)
        return w
    except ZeroDivisionError:
        return ":    (y + 1 = 0  34 + x = 0)"
    except Exception as e:
        return f"  : {e}"

#  2
def calculate_y(x: float):
    try:
        if math.sin(x) <= 0:
            raise ValueError("sin(x)   > 0   ln(sin(x))")
        y = math.exp(x) * math.sin(x) + math.log(math.sin(x))
        return y
    except ValueError as e:
        return f": {e}"
    except Exception as e:
        return f"  : {e}"

#  3
def power_operations(a: int):
    a2 = a * a  # 1- 
    a4 = a2 * a2  # 2- 
    a8 = a4 * a4  # 3- 
    a10 = a8 * a2  # 4- 

    return a8, a10

#  4
def swap_hundreds_tens(num: int) -> int:
    hundreds = num // 100
    tens = (num // 10) % 10
    ones = num % 10

    #     
    new_num = tens * 100 + hundreds * 10 + ones
    return new_num

#  5
def is_even(A: int) -> bool:
    #  ,    2  
    return A % 2 == 0



#  1
x, y = 2, -1
print(f" 1: w = {calculate_w(x, y)}")
x, y = 2, 3
print(f" 1: w = {calculate_w(x, y)}")

#  2
x_val = math.pi / 2
print(f" 2: y = {calculate_y(x_val)}")
x_val = 3 * math.pi / 2
print(f" 2: y = {calculate_y(x_val)}")

#  3
a = 2
a8, a10 = power_operations(a)
print(f" 3: a = {a}, a^8 = {a8}, a^10 = {a10}")

#  4
num = 123
print(f" 4:  : {num},  : {swap_hundreds_tens(num)}")

#  5
A = 7
print(f" 5: A = {A},  - {is_even(A)}")
```

```python
import math

#  1
def f(x: float) -> float:
    try:
        if x <= 3:
            return -x**2 + 3*x + 9
        else:
            return x / (x**2 + 1)
    except Exception as e:
        return f": {e}"

#  2
def is_isosceles(a: float, b: float, c: float) -> bool:
    try:
        # ,   
        if a + b <= c or a + c <= b or b + c <= a:
            raise ValueError("   !")
        #        
        return a == b or a == c or b == c
    except Exception as e:
        return f": {e}"

#  3
def convert_angle(value: float, mode: str) -> float:
    """
    mode = "rad_to_deg" ->   -> 
    mode = "deg_to_rad" ->   -> 
    """
    try:
        if mode == "rad_to_deg":
            return value * 180 / math.pi
        elif mode == "deg_to_rad":
            return value * math.pi / 180
        else:
            raise ValueError(" .  'rad_to_deg'  'deg_to_rad'.")
    except Exception as e:
        return f": {e}"


#  4
def bounding_rectangle(r1: tuple, r2: tuple) -> tuple:
    """
    r1 = (x1_left, y1_bottom, x1_right, y1_top)
    r2 = (x2_left, y2_bottom, x2_right, y2_top)
    : (x_left, y_bottom, x_right, y_top)
    """
    x_left = min(r1[0], r2[0])
    y_bottom = min(r1[1], r2[1])
    x_right = max(r1[2], r2[2])
    y_top = max(r1[3], r2[3])
    return (x_left, y_bottom, x_right, y_top)


#  5
def pawn_move(a: int, b: int, c: int, d: int) -> tuple:
    #  :     1 
    normal_move = (c == a and d == b + 1)
    #  :     1 
    capture_move = (d == b + 1 and (c == a - 1 or c == a + 1))
    return normal_move, capture_move



#  1
print(" 1:", f(2), f(5))

#  2
a, b, c = 5, 5, 7
print(f" 2:    ({a}, {b}, {c})  -", is_isosceles(a, b, c))
a, b, c = 1, 2, 10
print(f" 2:    ({a}, {b}, {c})  -", is_isosceles(a, b, c))

#  3
print(" 3: 180 =", convert_angle(180, "deg_to_rad"), "")
print(" 3:   =", convert_angle(math.pi, "rad_to_deg"), "")

#  4
rect1 = (0, 0, 2, 3)
rect2 = (1, 1, 4, 5)
print(" 4:   =", bounding_rectangle(rect1, rect2))

#  5
a, b, c, d = 4, 2, 5, 3
normal, capture = pawn_move(a, b, c, d)
print(f" 5:  ({a},{b}) -> ({c},{d}),   - {normal},  - {capture}")
```

```python
import math

#  1
def pythagoras_numbers(k: int):
    result = []
    for m in range(2, k):
        for n in range(1, k):
            if m > n:           #   
                a = m**2 - n**2
                b = 2 * m * n
                c = m**2 + n**2
                result.append((m, n, a, b, c))
    return result

#  2
def sum_of_sines(n: int, x: float) -> float:
    total = 0
    for i in range(1, n + 1):
        total += math.sin(x**i)
    return total

#  3
def series_sum(eps: float = 1e-3) -> float:
    total = 0.0
    n = 1
    while True:
        term = (2**n) / math.factorial(n - 1)
        total += term
        if term < eps:
            break
        n += 1
    return total

#  4
def has_positive(nums: list[int]) -> bool:
    return any(x > 0 for x in nums)



#  1
k = 10
print(" 1:   (m, n, a, b, c)")
for row in pythagoras_numbers(k):
    print(row)

#  2
n, x = 3, 1.0
print(f" 2:  = {sum_of_sines(n, x)}")

#  3
print(f" 3:   = {series_sum()}")

#  4
nums = [-5, -2, 0, 3]
print(f" 4:  {nums}   - {has_positive(nums)}")
```

```python
import math

#  1
def is_prime(num: int) -> bool:
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

def is_binary_palindrome(num: int) -> bool:
    binary_str = bin(num)[2:]   #    
    return binary_str == binary_str[::-1]

def prime_binary_palindromes(n: int) -> list[int]:
    return [x for x in range(2, n + 1) if is_prime(x) and is_binary_palindrome(x)]


#  2
def sum_three_elements(D: list[int], k: int) -> int:
    """
          ,
       k (  1).
    """
    try:
        if k < 1 or k + 2 > len(D):
            raise ValueError("     ")
        return D[k - 1] + D[k] + D[k + 1]
    except Exception as e:
        return f": {e}"


#  3
def factorial_sum_even(n: int) -> int:
    try:
        if n % 2 != 0:
            raise ValueError("n   ")
        if n > 20:
            raise ValueError("n    20")

        total = 0
        for i in range(2, n + 1, 2):
            total += math.factorial(i)
        return total
    except Exception as e:
        return f": {e}"


#  1
n = 100
print(" 1:   <= n   :", prime_binary_palindromes(n))

#  2
D = [1, 2, 3, 4, 5, 6]
print(" 2:")
print("D[1]+D[2]+D[3] =", sum_three_elements(D, 1))
print("D[3]+D[4]+D[5] =", sum_three_elements(D, 3))
print("D[4]+D[5]+D[6] =", sum_three_elements(D, 4))
print("D[5]+D[6]+D[7] =", sum_three_elements(D, 5))

#  3
n = 10
print(f" 3:    {n}! =", factorial_sum_even(n))
n = 21
print(f" 3:    {n}! =", factorial_sum_even(n))
```

```python
#  1
def find_positive_pairs(arr: list[int]):
    first, last = None, None
    for i in range(len(arr) - 1):
        if arr[i] > 0 and arr[i + 1] > 0:
            if first is None:
                first = (i + 1, i + 2)  #   1
            last = (i + 1, i + 2)
    return first, last


#  2
def even_index_elements(arr: list[int]) -> list[int]:
    #     2
    return arr[1::2]


#  3
def cyclic_shift_left(arr: list[int]) -> list[int]:
    if not arr:
        return arr
    return arr[1:] + arr[:1]


#  4
def count_multiples_of_7(matrix: list[list[int]]) -> int:
    count = 0
    for row in matrix:
        for elem in row:
            if elem % 7 == 0:
                count += 1
    return count


#  1
arr1 = [1, -2, 3, 5, -1, 7, 8]
first, last = find_positive_pairs(arr1)
print(" 1:  =", arr1)
print(" :", first)
print(" :", last)

#  2
arr2 = [10, 20, 30, 40, 50, 60]
print(" 2:  =", arr2)
print("   :", even_index_elements(arr2))

#  3
arr3 = [1, 2, 3, 4, 5]
print(" 3:  =", arr3)
print("  :", cyclic_shift_left(arr3))

#  4
matrix = [
    [7, 14, 3],
    [21, 8, 28],
    [1, 2, 3]
]
print(" 4:  =", matrix)
print(" ,  7:", count_multiples_of_7(matrix))
```

```python
import re

#  1
def unique_words(sentence: str) -> set[str]:
    #     (    )
    words = re.findall(r"\w+", sentence.lower())  #  
    return set(words)


#  2
def longest_a_sequence(s: str) -> int:
    max_len, current = 0, 0
    for ch in s:
        if ch == "":  #   ""
            current += 1
            max_len = max(max_len, current)
        else:
            current = 0
    return max_len


#  3
def unique_chars(s: str) -> str:
    result = []
    for ch in s:
        if s.count(ch) == 1:
            result.append(ch)
    return "".join(result)



#  1
sentence = ", !    ."
print(" 1:  :", unique_words(sentence))

#  2
s1 = "   "
print(" 2:   '' =", longest_a_sequence(s1))

#  3
s2 = ""
print(" 3: ,   :", unique_chars(s2))
```

```python
import re

#  1
def task1_write_and_find(filename: str, numbers: list[float]) -> float:
    #    (    )
    with open(filename, "w", encoding="utf-8") as f:
        for num in numbers:
            f.write(f"{num}\n")

    #          
    max_abs = float("-inf")
    with open(filename, "r", encoding="utf-8") as f:
        nums = [float(line.strip()) for line in f]
        for i in range(0, len(nums), 2):
            max_abs = max(max_abs, abs(nums[i]))
    return max_abs

#  2
def task2_find_expensive_constructor(filename: str) -> float:
    max_cost = 0
    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split(";")
            if len(parts) < 4:
                continue
            name, cost, age_min, age_max = parts
            cost = float(cost)
            if "" in name.lower():
                max_cost = max(max_cost, cost)
    return max_cost


#  3
def task3_extract_expressions(filename: str) -> list[tuple[str, float]]:
    results = []
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read()

    #  :   
    pattern = r"(\d+(?:\.\d+)?)\s*([+\-*/])\s*(\d+(?:\.\d+)?)"
    matches = re.findall(pattern, text)

    for a, op, b in matches:
        a, b = float(a), float(b)
        if op == "+":
            val = a + b
        elif op == "-":
            val = a - b
        elif op == "*":
            val = a * b
        elif op == "/":
            val = a / b if b != 0 else float("inf")
        results.append((f"{a} {op} {b}", val))
    return results



#  1
nums = [1.5, -3.2, 7.1, -8.4, 2.0]
print(" 1:", task1_write_and_find("numbers.txt", nums))

#  2
print(" 2:", task2_find_expensive_constructor("Assort.txt"))

#  3
print(" 3:", task3_extract_expressions("exp.txt"))
```
